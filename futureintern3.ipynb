# checking the data wheather it is normal distribution or not using histogram and box plot after using standardization *technique*

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)


X_scaled_df = pd.DataFrame(X_scaled, columns=x.columns)


plt.figure(figsize=(12, 8))
for i, column in enumerate(X_scaled_df.columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(X_scaled_df[column], kde=True)
    plt.title(f'Histogram of standardized {column}')

plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 8))
for i, column in enumerate(X_scaled_df.columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=X_scaled_df[column])
    plt.title(f'Box plot of standardized {column}')

plt.tight_layout()
plt.show()
```
